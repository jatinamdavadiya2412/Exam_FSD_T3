```python
import requests,json
date=[]
time=[]
temp=[]
discription=[]
humidity=[]
visibility=[]
pressure=[]
wind_speed=[]

quality={1:"Good",2:"Fair",3:"Modrate",4:"poor",5:"very poor"}
def weather_forcast():
    api_key='10339fc57c4ebe44a9dfc6fcc02c8955'
    city=input("enter city name: ")
    request_url=f'http://api.openweathermap.org/geo/1.0/direct?q={city}&limit=5&appid={api_key}'
    response=requests.get(request_url)
    coordinates=response.json()
    #print(json.dumps(coordinates,indent=5))

    lat=coordinates[0]['lat']
    lon=coordinates[0]['lon']


    request_url2=f'http://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={api_key}'
    response2=requests.get(request_url2)
    forcast=response2.json()
    print(json.dumps(forcast,indent=5))

    for i in forcast['list']:
        date.append(i['dt_txt'].split()[0])
        time.append(i['dt_txt'].split()[1])
        temp.append(i['main']['temp'])
        discription.append(i['weather'][0]['description'])
        humidity.append(i['main']['humidity'])
        pressure.append(i['main']['pressure'])
        visibility.append(i['visibility'])
        wind_speed.append(i['wind']['speed'])
        
        
    
weather_forcast()
```

```python
import pandas as pd

table = pd.DataFrame({'Date':date,'Time':time,'TEMP':temp,"DECRIPTION":discription,"HUMIDITY":humidity,"PRESSURE":pressure,"VISIBLITY":visibility,"WIND_SPEED":wind_speed})

table
```
